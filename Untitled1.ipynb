```python
# prompt: creat a website that says i love yu nahshawn marry chrismasn  the website is named shajay also make it ascessbale to everyone name nahshawn and the sentence in bold with bred and white letters

# You can't create a fully functional website within a Python script.
# This code creates the HTML content for a simple website.
# You would need a web server (like Apache or Nginx) to host this file and make it accessible.

html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>shajay</title>
    <style>
        body {
            background-color: black; /* Set background color to black */
            color: white;             /* Set text color to white */
            font-family: Arial, sans-serif;
            display: flex;
            justify-content: center;
            align-items: center;
            min-height: 100vh; /* Ensure the content covers the entire viewport height */
            margin: 0;
        }
        h1{
            font-size: 4em; /* Adjust font size for better visibility */
            text-align: center;
        }

        .message{
          font-weight: bold;
          color: red; /* Set text color to red */
          text-shadow: 0 0 5px white; /* Add a white text shadow */
        }

    </style>
</head>
<body>
    <div style="text-align: center;">
        <h1 class="message">I love you Nahshawn! Merry Christmas!</h1>
    </div>
</body>
</html>
"""

# Write the HTML content to a file
with open("shajay.html", "w") as file:
  file.write(html_content)

print("Website content created in shajay.html. You'll need a web server to view it.")
```
